###### Countplot using seaborn to get count without label

###### Countplot using seaborn to get count with labels

```python
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x='State', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc = {'figure.figsize':(15,4)})
```

```python
ax = sns.countplot(x='Orders', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x='Age Group', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

###### plotting a bar chart for gender vs total amount

```python
data_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='Gender', y='Amount', data = data_gen)
sns.set(rc={'figure.figsize':(5, 3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

###### barplot with rotated label text

```python
data_gen1 = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
ax = sns.barplot(x='State', y='Orders', data = data_gen1, hue = 'State')
sns.set(rc = {'figure.figsize':(10,5)})
ax.tick_params(axis='x', rotation=-45)
```

```python
data_gen2 = df.groupby(['Age Group'], as_index=False)['Marital_Status'].sum().sort_values(by='Marital_Status', ascending=False)
axx = sns.barplot(x='Age Group', y = 'Marital_Status', data = data_gen2, hue = 'Age Group')
sns.set(rc = {'figure.figsize':(10, 7)})
for bars in axx.containers:
    axx.bar_label(bars)
```

```python
data_gen2 = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
axxx = sns.barplot(x='Gender', y='Amount', data = data_gen2, hue = 'Gender')
sns.set(rc= {'figure.figsize': (10,5)})
for bars in axxx.containers:
    axxx.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue = 'Gender')
```
